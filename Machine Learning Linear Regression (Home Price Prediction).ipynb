```python
df = df.drop(columns=["Address","Date","Postcode","YearBuilt","Lattitude","Longtitude"])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df[["Propertycount","Distance","Bedroom2","Bathroom","Car"]] = df[["Propertycount","Distance","Bedroom2","Bathroom","Car"]].fillna(0)
```

```python
df[["Landsize","BuildingArea"]] = df[["Landsize","BuildingArea"]].fillna(df[["Landsize","BuildingArea"]].mean())
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df["Method"].unique()
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
x = df.drop(columns=["Price"])
```

```python
x.head()
```

```python
y = df[["Price"]]
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```
